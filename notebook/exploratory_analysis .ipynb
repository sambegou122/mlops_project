```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data 

```python
train = pd.read_csv('data/train.csv', delimiter=',')
test = pd.read_csv('data/test.csv',  delimiter=',')
valid = pd.read_csv('data/valid.csv',  delimiter=',')
```

```python
train.drop(['Unnamed: 0', "film-url"], axis=1, inplace=True)
test.drop(['Unnamed: 0', "film-url"], axis=1, inplace=True)
valid.drop(['Unnamed: 0', "film-url"], axis=1, inplace=True)
```

## Data Exploration
We have 3 datasets:
- **train.csv**: the training set
- **test.csv** the test set
- **valid.csv**: For the validation step data

In below we will explore we can see that:

- All those datasets have the same structure and the ar'nt any missing values.

- The porportion of the target variable is ~0.5 which means that we have a balanced dataset for all datasets.

- The distribution of the words in the text for the 3 datasets is almost the same.

```python
train.isna().sum(), test.isna().sum(), valid.isna().sum()
```

```python

print(f"Proportion of the target in train {train['polarity'].value_counts()/len(train)}")
print(f"Proportion of the target in test {test['polarity'].value_counts()/len(test)}")
print(f"Proportion of the target in valid {valid['polarity'].value_counts()/len(valid)}")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
train['polarity'].value_counts().plot(kind='bar', title='target distribution', ax=ax1)
test['polarity'].value_counts().plot(kind='bar', title='target distribution', ax=ax2, )
valid['polarity'].value_counts().plot(kind='bar', title='target distribution', ax=ax3,)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
train['review'].apply(lambda x: len(x.split(" "))).hist(bins=100, figsize=(15, 5), ax=ax1)
ax1.set_title('distribtion of train review word length in train')
test['review'].apply(lambda x: len(x.split(" "))).hist(bins=100, figsize=(15, 5), ax=ax2)
ax2.set_title('distribtion of train review word length in test')
valid['review'].apply(lambda x: len(x.split(" "))).hist(bins=100, figsize=(15, 5), ax=ax3)
ax3.set_title('distribtion of train review word length in valid')
```
